## convert to classification task

```python
clean_data = data.copy()
clean_data['high_petal_length'] = (clean_data['petal_length']>5)*1
clean_data['high_petal_length']
```

using petal and sepal lengths and wdth to prdict the species.

```python
lengths=['petal_length','petal_width','sepal_length','sepal_width']
```

```python
x = clean_data[lengths].copy() #if complicated values use deep copy i.e in copy(deep=True).
```

```python
x.columns
```

```python
x.head()
```

```python
y=clean_data['species']
y.head()
```

## perform test and train split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=324)
```

```python
#print(x_train)
#print(x_test)
#print(y_train)
#print(y_test)
#x_train.head()
#y_train.describe()
```

### fit on train set

```python
species_classifier = DecisionTreeClassifier(max_leaf_nodes=5,random_state=0)
species_classifier.fit(x_train, y_train)
```

```python
type(species_classifier)
```

### predict on test data

```python
predictions = species_classifier.predict(x_test)
```

```python
predictions[:10]
```

```python
y_test['species'][:10]
```

### measure accuracy of the classifier

```python
accuracy_score(y_true=y_test, y_pred=predictions)
```
