Correlation

Now we'll try to find which features are strongly correlated with SalePrice. We'll store them in a var called golden_features_list. We'll reuse our df_num dataset to do so.

```python
df_num_corr = df_num.corr()['Item_Outlet_Sales'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Item_Outlet_Sales:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
def sca_pl(X,Y):
    plt.scatter(X,Y)
    plt.show()
    plt.clf()
    plt.title('scatter plot of X and Y')
    plt.xlabel('x-axis')
    plt.ylabel('Item_Outlet_Sales')
    return X, Y
Y = df['Item_Outlet_Sales']
sca_pl(df['Item_Weight'], Y) 
```
